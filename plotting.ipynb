```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('timings.csv')
```

```python
#create another column in the df that is the sqrt_size column squared
data = data[data['sqrt_size'] < 1000]
data['size'] = data['sqrt_size']**2
```

```python
for a in data['algorithm'].unique():
    tmp_data = data[data['algorithm'] == a].groupby('sqrt_size')['time'].mean()
    plt.plot(tmp_data.index, tmp_data.values, label=a)

plt.legend()
plt.xticks(data['sqrt_size'].unique(), rotation=45)
plt.xlabel("Maze size (sqrt n)")
plt.ylabel("Execution time (s)")
plt.title("Execution time of maze generation algorithms")
```
